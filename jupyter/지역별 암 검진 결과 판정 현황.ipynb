```python
import numpy as np
import pandas as pd
```

```python

#=======================================================================================
#이상있음 : 양성질환, 위암의심, 위암, 기타, 기존 암환자
#이상없음 : 이상소견없음


data1 = pd.read_csv('../data/지역별 위암검진 결과 판정 현황.csv', encoding='cp949')
result1 = data1.iloc[1:].replace('-','0')
result1=result1[result1['성별(1)']=='합계'].reset_index()

result1['이상있음']=result1['2019.2'].astype(float)+result1['2019.3'].astype(float)+result1['2019.4'].astype(float)+result1['2019.5'].astype(float)+result1['2019.6'].astype(float)
result1['이상없음']=result1['2019.1'].astype(float)

# =======================================================================================
#이상있음 : 잠혈반응 있음, 대장폴립, 대장암의심, 대장암, 기타, 기존 암환자
#이상없음 : 잠혈반응 없음, 이상소견 없음

data2 = pd.read_csv('../data/지역별 대장암 검진 결과 판정현황.csv', encoding='cp949')
result2 = data2.iloc[1:].replace('-','0')
result2=result2[result2['성별(1)']=='합계'].reset_index()

result2['이상있음']=result2['2019.2'].astype(float)+result2['2019.4'].astype(float)+result2['2019.5'].astype(float)+result2['2019.6'].astype(float)+result2['2019.7'].astype(float)+result2['2019.8'].astype(float)
result2['이상없음']=result2['2019.1'].astype(float)+result2['2019.3'].astype(float)

#=======================================================================================
#이상있음 : 양성질환, 유방암 의심, 기존 암환자
#이상없음 : 정상, 판정유보

data3 = pd.read_csv('../data/지역별 유방암 검진 결과 판정 현황.csv', encoding='cp949')
result3 = data3.iloc[1:].replace('-','0').reset_index()

result3['이상있음']=result3['2019.2'].astype(float)+result3['2019.3'].astype(float)+result3['2019.5'].astype(float)
result3['이상없음']=result3['2019.1'].astype(float)+result3['2019.4'].astype(float)

#=======================================================================================
#이상있음 : 반응성 소견 및 감염성 질환, 비정형 세포이상, 자궁경부암 전구단계 의심, 자궁경부암 의심, 기타, 기존 암환자
#이상없음 : 이상소견 없음


data4 = pd.read_csv('../data/지역별 자궁경부암 검진 결과 판정 현황.csv', encoding='cp949')
result4 = data4.iloc[1:].replace('-','0').reset_index()

result4['이상있음']=result4['2019.2'].astype(float)+result4['2019.3'].astype(float)+result4['2019.4'].astype(float)+result4['2019.5'].astype(float)+result4['2019.6'].astype(float)+result4['2019.7'].astype(float)
result4['이상없음']=result4['2019.1'].astype(float)

#=======================================================================================
#이상있음 : 추적검사 요망, 간암의심, 기타, 기존 암환자
#이상없음 : 간암의심 소견없음, 


data5 = pd.read_csv('../data/지역별 간암 검진 결과 판정 현황.csv', encoding='cp949')
result5 = data5.iloc[1:].replace('-','0')
result5=result5[result5['성별(1)']=='합계'].reset_index()

result5_1=result5[result5['반기별(1)']=='상반기'].reset_index()
result5_2=result5[result5['반기별(1)']=='하반기'].reset_index()

result5_total=pd.DataFrame()
result5_total['시군구별(1)']=result5_1['시군구별(1)']
result5_total['시군구별(2)']=result5_1['시군구별(2)']
result5_total['2019']=result5_1['2019'].astype(float)+result5_2['2019'].astype(float)
result5_total['2019.1']=result5_1['2019.1'].astype(float)+result5_2['2019.1'].astype(float)
result5_total['2019.2']=result5_1['2019.2'].astype(float)+result5_2['2019.2'].astype(float)
result5_total['2019.3']=result5_1['2019.3'].astype(float)+result5_2['2019.3'].astype(float)
result5_total['2019.4']=result5_1['2019.4'].astype(float)+result5_2['2019.4'].astype(float)
result5_total['2019.5']=result5_1['2019.5'].astype(float)+result5_2['2019.5'].astype(float)

result5_total['이상있음']=result5_total['2019.2']+result5_total['2019.3']+result5_total['2019.4']+result5_total['2019.5']
result5_total['이상없음']=result5_total['2019.1']

#=======================================================================================
data6 = pd.read_csv('../data/지역별 암검진 대상 및 수검인원 현황.csv', encoding='cp949')
result6 = data6.iloc[2:].replace('-','0')
result6=result6[result6['성별(1)']=='합계'].reset_index()
result6['수검율']=(result6['2019.3'].astype(float)+result6['2019.5'].astype(float)+result6['2019.7'].astype(float)+result6['2019.9'].astype(float)+result6['2019.11'].astype(float))/(result6['2019.2'].astype(float)+result6['2019.4'].astype(float)+result6['2019.6'].astype(float)+result6['2019.8'].astype(float)+result6['2019.10'].astype(float))

#=======================================================================================
new_data=pd.DataFrame()
new_data[0]=result6['시군구별(1)']
new_data[1]=result6['시군구별(2)']
new_data[2]=result6['수검율']
new_data[3]=(result1['이상없음']+result2['이상없음']+result3['이상없음']+result4['이상없음']+result5_total['이상없음'])/(result6['2019.3'].astype(float)+result6['2019.5'].astype(float)+result6['2019.7'].astype(float)+result6['2019.9'].astype(float)+result6['2019.11'].astype(float))
new_data[4]=(result1['이상있음']+result2['이상있음']+result3['이상있음']+result4['이상있음']+result5_total['이상있음'])/(result6['2019.3'].astype(float)+result6['2019.5'].astype(float)+result6['2019.7'].astype(float)+result6['2019.9'].astype(float)+result6['2019.11'].astype(float))
new_data=new_data[new_data[1] != '소계']


new_data.to_csv('result/processed_지역별 암 검진 결과 판정 현황.csv', encoding='cp949')
```
