```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_satisfaction = pd.read_csv("../data/지역별 의료서비스만족도(의원).csv", encoding='cp949')
df_satisfaction
```

```python
result = df_satisfaction[df_satisfaction["행정구역(시도)별(1)"] != "전국"]
result
```

```python
result = result[result["특성별(1)"] == "전체"]
result
```

```python
result['만족 점수'] = (result['매우 만족'].astype(float)  * 0.01) + (result['약간 만족'].astype(float)  * 0.0075)+ (result['보통'].astype(float)  * 0.005) + (result['약간 불만족'].astype(float)  * 0.0025)
```

```python
result
```

```python
score_result = result[['행정구역(시도)별(1)', '만족 점수']]
```

```python
score_result
```

```python
score_result.to_csv("./result/processed_지역별 의료서비스만족도(의원).csv", header=False, index=False, encoding='cp949')
```
