```python
import pandas as pd
import numpy as np
```

```python
file = "../data/지역별 교육수준.csv"

df = pd.read_csv(file, encoding="cp949")
df = df.fillna(0)
print(df['행정구역별(시군구)'].iloc()[0])
```

```python
area_list = []
result = []
count = 0
temp = []
for line in df.iloc():
    if temp == []:
        temp.append(line[0])
        temp.append(line[5])
    elif line[0] == temp[0]:
        temp.append(line[5])
    elif df['행정구역별(시군구)'].iloc()[count-1] == temp[0]:
        result.append(temp)
        temp = []
        temp.append(line[0])
        temp.append(line[5])
        
    count += 1
    
```

```python
n_df = pd.DataFrame(result)
print(n_df)
```
