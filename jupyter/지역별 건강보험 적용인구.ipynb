```python
df_population = pd.read_csv("../data/지역별 인구수.csv", encoding='cp949')
df_population
```

```python
df_insurance = pd.read_csv("./result/processed_지역별 건강보험 적용인구.csv", encoding='cp949')
df_insurance
```

```python
df_bind = df_insurance
df_bind['인구수'] = df_population['인구수']
```

```python
df_bind['비율'] = df_bind['적용인구'] / df_bind['인구수'] * 100
df_bind
```

```python
max_rate = df_bind['비율'].max()
min_rate = df_bind['비율'].min()
print(max_rate, min_rate)
```

```python
df_bind['정규분포'] = df_bind['비율'].astype(float) - df_bind['비율'].min()
df_bind
```

```python
df_bind['정규분포'] = df_bind['정규분포'] / df_bind['정규분포'].max()
df_bind['정규분포'].max(), df_bind['정규분포'].min()
df_bind
```

```python
df_bind.drop(['비율', '적용인구', '인구수'], axis=1, inplace=True)
df_bind
```

```python
df_bind.to_csv("./result/processed_지역별 건강보험 적용인구.csv", header=False, index=False, encoding='cp949')
```
